The final step in the cleaning process is synthesizing the final cleaned dataset that will be used in the analysis and re-expressing categorical variables as numerical data types. The only variables required for my cleaned dataframe are:

Categorical variables that have more that just binary responses will need to be re-expressed using one hot encoding. This process creates dummy variables to store a binary indicator for each data entry, isolating markers for a project's category or country; for example a project designated in the Comics category will have a '1' in the main_category_Comics column but a zero in all the other category columns. This technique of re-expressing categorical data as dummy variables is advantageous as it will allow for the representation of multiple groups within a single regression equation. However, the drawback is that the creation of so many dummy variable columns results in poor performance as the regression model's workload is increased (Blankmeyer, 2006). 

Given the limited amount of data on projects made outside of the US, creating dummy variables for every country code may result in redundancy in the model per the drawback mentioned above. As such, it would be more preferrable to re-express the  data within the Country column in a binary format: with projects based in the US being denoted as '1', and projects made outside the US denoted as '0'. The benefit to doing so will be a reduction of bloat within the dataframe and computational load on the upcoming analysis, but the downside is a loss of granularity in knowing exactly which countries may have statistical significance on a successful project outcome.

When using one hot encoding, it is important to ensure that keeping all of the dummy variable columns will result in perfect multicollinearity. As such, the first column created during the one hot encoding process will be dropped.

```python
#dropping unnecessary columns from dataframe
cleandf = df.drop(columns = ["category", "currency", "deadline", "launched"])

#re-expressing categorical variables as numerical through replacement of success with 1 and failure with 0
cleandf["state"] = df["state"].replace({"successful": 1, "failed": 0})

#re-expressing categorical variables as numerical through dummy variables, ensuring that the first variable is dropped 
#to prevent multicollinearity
cleandf = pd.get_dummies(cleandf, columns=["main_category"], drop_first=True)

#re-expressing variable within the country column in binary format. US = 1, non-US = 0
cleandf['country'] = (cleandf['country'] == 'US').astype(int)

#renaming the country column to is_us
cleandf.rename(columns={'country':'is_us'}, inplace=True)
```

The final step in preparing the dataset for analysis is the standardization of data. This technique takes continuous data that ranges on a numeric scale with no upper boundary, and scales it along the z-score; re-expressing each value as a measure of standard deviations away from the variable's mean. This technique takes multiple sets of data, i.e. the continuous variables within the dataset and scales them to an even number range, and this technique was selected for the purpose of the Logistic Regression analysis to ensure no single feature dominates the model's ability to assess influence of the campaign's outcome, and it helps with interpretability of relative variable importance when assessing the regression model's coefficients. Furthermore, this technique is advantageous for use with Logistic Regression as LR model algorithms have increased efficiency in performance when the range of numerical data inputs is standardized (Khoong, 2023). One drawback to the use of this technique would be that the raw data becomes less interpretable at a glance; assessing a funding goal of $40,000 makes more sense than assessing a scaled metric of 3.68. However as the continuous data has already been reviewed above, this will not be an issue.

```python
#creating dataframe to house data for z-scaling
scaled_df = cleandf

#conducting z-scaling on columns containing continuous data
scaled_df["usd_goal_real"] = ((cleandf["usd_goal_real"] - cleandf["usd_goal_real"].mean())
                                / cleandf["usd_goal_real"].std())

scaled_df["duration"] = ((cleandf["duration"] - cleandf["duration"].mean())
                                / cleandf["duration"].std())

scaled_df["backers"] = ((cleandf["backers"] - cleandf["backers"].mean())
                                / cleandf["backers"].std())

#printing final cleaned dataframe to observe changes
scaled_df
```

As can be seen from the final cleaned dataframe above, all variables are currently expressed as numerical data for appropriate analysis. From all the cleaning performed, the dataset was reduced from its original 378,661 entries by approximately 110,000, with 268,361 rows remaining. Finally, the dataframe will be exported into a new csv file and submitted alongside the analysis.

```python
#exporting dataframe into csv file
scaled_df.to_csv('cleaned_dataset.csv', index=False)
```

## Section D: Analysis
### Parametric Testing
Kickstarter officially states that projects that lasted longer than 60 days generally did not see success, and that they instead suggest running a campaign for less than 30 days to increase the likelihood of meeting the funding goal (Kickstarter, 2024). To test this statement's validity, a t-test on the distribution of project duration between those that have Succeeded and those that have Failed can be used. A t-test is a statistical test that is utilized to assess the statistical significance of differences between means of two populations. The reason why a t-test was chosen to evaluate the means of duration for successful and unsuccessful projects is because hypothetically, this test would provide insight on whether differences between the two populations are resultant from statistically different distributions of project durations, assuming that successful projects have a distribution of duration lower than those that have failed. As it is a parametric test, it makes the same assumptions as other parametric tests, primarily that the data is normally distributed. The major benefit of a t-test is that it is very straightforward and simple to calculate, but the drawback is that t-tests are very sensitive to their assumptions; violations of the normality assumption especially tends to introduce skew and can lead to inaccurate results (360Studies, 2023). 

For this study, both a Q-Q plot and a Shaprio-Wilk test were selected as they can be utilized to test the normality of the data. A Q-Q (quantile-quantile) plot is a scatterplot used for assessing normality of a population's distribution. The benefit of using this plot is due to the visual representation of normality through comparison against a reference line. The better the fit of the data against the line, the greater the likelihood of the data being normally distributed. The downside of q-q plots is that there is no defined metric to determine how *poor* the fit against the reference line must be before normality is deemed to be non-existent within the data.

Consider the below:

```python
#creating a qqplot through the statsmodel library, using data scaled through standardization
sm.qqplot(scaled_df["duration"], line='45')
plt.show()
```

From the Q-Q plot above, one can see that the data touches the red line but does not entirely conform to it, indicating that the data does not present a strong distribution of normality. As mentioned above, it is difficult to quantify exactly *how* much the data should stray from the line before it can be deemed to not have a normal distribution. As such, we can take a deeper look through a Shapiro-Wilk test. 

A Shapiro-Wilk (S-W) test is a statistical test that assesses the normal distribution of a dataset, but unlike a q-q plot, there is no visual representation of data *fitting* the ideal distribution. The test instead establishes a null hypothesis that the data *is* normally distributed, and provides a statistical metric, giving an advantage through evaluation of a numerical p-value for a more approachable assessment of normality. If the p-value of the S-W test is below an established alpha of 0.05, the null hypothesis is rejected, as there is not enough evidence to indicate that the data *is* normally distributed. It is important to note that Shapiro-Wilk tests are strongly sensitive to large datasets (Rehal, 2023), which is the major disadvantage in using the test for this specific assessment.

```python
#determining the p-value through a shapiro-wilk test
stats.shapiro((scaled_df["duration"]))
```

As the p-value of the S-W test is less than 0.05, we reject the null-hypothesis that the data is normally distributed. Based on these circumstances, the data would normally be invalid for a t-test as it invalidates the assumption of normal distribution. However, the Central Limit Theorem states that the distribution of a sample approaches an approximate normal distribution as the sample size of data becomes larger, even if the population isnt normally distributed (Turney, 2023). 

Consider two samples taken from the 'duration' population: projects that have succeeded and projects that have failed. Given the large quantity of observable data within the dataset, it can be assumed that these samples from the population of the 'duration' variable are normally distributed based on the central limit theorem, satisfying the normality requirement for a t-test.

In a t-test, the null hypothesis is that population of dataset 1 is equal to the population of dataset 2. Thus, if the resultant p-value is below 0.05, then there is sufficient evidence to accept the alternative hypothesis that the two populations are statistically different from one another. To conduct the t-test, the scipy.stats library comes with a built in t-test function that requires a few things:
- Two datasets
- Equivalence of variance between the two datasets (equal_var)

Equivalence of variance will be calculated using the np.var function from the numpy library. If the resulting ratio between the two variances is less than 4:1, the given datasets are considered to have equal variances (Geeksforgeeks, 2022).

```python
#establishing data groups for t-test based on project state (Geeksforgeeks, 2022)
data_group1 = df.loc[scaled_df["state"] == 1, "duration"]
data_group2 = df.loc[scaled_df["state"] == 0, "duration"]

#printing the variances of the datasets
print(np.var(data_group1), np.var(data_group2))
```

```python
#as the ratio between the variances is less than 4:1, variances are considered to be equal
stats.ttest_ind(a=data_group1, b=data_group2, equal_var=True)
```

As the p-value of the t-test is less than 0.05, there is sufficient evidence to reject the null hypothesis, concluding that the duration of a project's campaign does have influence on the outcome of the project as the mean duration between both population samples are statistically different. 

### Logistic Regression
Logistic regression models are statistical models that are utilized primarily for classification and prediction analysis by determining relationships between several independant variables and a dependant categorical variable. The model develops an understanding of these relationships by training on observed data, resulting in a model that can predict a dependant variable's likely outcome based on data it has analyzed (Kanade, 2022). 

Logistic Regression modelling was chosen for this study as the goal is to assess a binary outcome of project success/failure based on independant variables, which aligns perfectly with the dichotomous predictive capabilities described above. Furthermore, a logistic regression model provides insight into the likelihood of an outcome through each variable's coefficient: Unlike other forms of regression, logistic models are not predicting a unit of measurement on a linear scale, they are instead predicting the probability of an outcome of 'success' which either increases or decreases per 1 unit of measurement added for each variable. The coefficients of the model directly indicate the magnitude of which they influence outcome probability and are utilized to construct an equation for the log-odds of a successful outcome for a campaign.

Logistic regression models are advantageous for use in this study as training the model does not require significant computational power to perform, making it an efficient choice for analysis even with the large dataset being utilized. The downside of logistic regression is that predictive accuracy is reduced with smaller sample sizes, but as the dataset utilized for this study is large this will not be an issue.

### Justification of Reduction
One of the assumptions of logistic regression requires there to be no multicollinearity between the independant variables. As the model is attempting to determine relationships between each predictor variable and the dependant variable, highly correlated predictor variables can cause confusion as the model cannot assess each variable's influence on the outcome independantly (Kanade, 2022). As such, any variables that demonstrate multicollinearity with another must be removed in order for the predictive model to be effective.

I have chosen to utilize the Variance Inflation Factor (VIF) of the dataset's features to reduce multicollinearity between my independant variables as VIF values over 10 indicate a strong correlation between the independant variables. Upon running an analysis of the VIF of each independant variable, any feature with the highest VIF will be removed and the VIF will be run again, repeating until there are no variables with a VIF above 10 (Keith, 2021). This technique was chosen as it is a simple iterative process, being able to easily identify at a glance the variables with significant multicollinearity which is also its greatest advantage. There is however a disadvantage to using VIF for assessing multicollinearity and that is that this technique typically only assesses the paired correlation between predictors and has difficulty in capturing higher-dimensional multicollinearity (FasterCapital, 2024).

```python
#Conducting VIF analysis with all variables except the dependant variable (Keith, 2021)
X = scaled_df.drop(columns=["state"])

vif_df = pd.DataFrame()
vif_df["feature"] = X.columns

vif_df["VIF"] = [variance_inflation_factor(X.values, i)
for i in range(len(X.columns))]
print(vif_df)
```

As there are no VIF values above 10, there is no feature required to be removed as the multicollinearity assumption is satisfied. From here, the logistic regression model can be built using the statsmodel library. In optimizing the model's performance, additional feature reduction can be performed. To do so, backwards stepwise elimination will be conducted by removing the least important variable from the model; determined through an established alpha level of 0.05. 

Much like the feature reduction method proposed for VIF analysis the model will be run using all the features within the scaled_df dataset, the variable with the highest p-value over 0.05 will be removed and the process will be repeated until there are no independant variables within the model that yield a p-value over 0.05. The justification and the advantage for using this reduction technique is aimed at attaining a model solely containing independant variables that demonstrate significant influence on project success. However, one disadvantage of using this method is that dropped variables will not be re-entered into the analysis (Ziaul et al, 2020).

### Model

```python
#Creating logistic regression model (Li, 2017)
lr_model = sm.Logit(scaled_df["state"], scaled_df.drop(columns=["state"]).assign(const=1))
lr_result = lr_model.fit()
print(lr_result.summary())
```

No additional feature elimination is needed as all the independant variables listed have a p-value less than 0.05, demonstrating statistical significance in influencing the dependant variable's outcome. This is further supported by the model's LLR p-value. As the value is below 0.05, this model is deemed to carry statistical significance to the prediction of the dependant variable.

### Confusion Matrix
To assess the accuracy of the predictive model, the dataset will be split into training and testing sets at a 70/30 split. The former set will provide observed campaigns for the model to train on, and predicted results can be plotted against test data to verify the True Positives and True Negatives the model was able to predict.

The confusion matrix was selected as a means of assessing accuracy of the predictive model due to the benefit of its comprehensive visual and numerical presentation, making it easy to identify quantity of accurate predictions through a heatmap when generated through the seaborn library. The main disadvantage of using a heatmap is that it would become significantly more confusing to interpret if it had to predict an outcome of more than two classes, however this is not a problem for this study.

The below is the creation of the confusion matrix for the logistic regression model, alongside its accuracy calculation determined through a comparison of successes/failures predicted by the LR model and observed successes/failures within the testing dataset (W3Schools, 2024).

```python
#establishing the variables 
X = scaled_df.drop(columns=["state"]).assign(const=1)
y = scaled_df["state"]

#creating confusion model with accuracy (W3Schools, 2024)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify = y)
logregmodel = LogisticRegression()
logregmodel.fit(X_train, y_train)
y_pred = logregmodel.predict(X_test)
print('Accuracy of logistic regression model: '+ str((logregmodel.score(X_test, y_test))))
lrconfusion_matrix = confusion_matrix(y_test, y_pred)

#plotting the results from the confusion matrix
cm_display = ConfusionMatrixDisplay(confusion_matrix = lrconfusion_matrix, display_labels = [0, 1])
cm_display.plot()
plt.show() 
```

As can be seen from the above, the confusion matrix generated for the Logistic Regression model has predicted 47,144 true negatives and 25,934 true positives, which were correct predictions. Conversely, the model predicted 4,939 false negatives and 2,492 false positives, which were incorrect predictions. Overall, the model has a very strong level of accuracy in its predictions, given the accurate predictions made as well as the calculated accuracy score of 90.7%.
